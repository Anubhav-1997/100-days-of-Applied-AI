```
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/1331668.jpg',0)
plt.imshow(img)
plt.show()

kernel = np.ones((5,5),np.uint8)
```

# **Erosion Function**

```
erosion = cv2.erode(img,kernel,iterations = 1)
plt.xticks([]), plt.yticks([])
plt.imshow(erosion)
plt.show()
```

## **Dilation Function**

```
dilation = cv2.dilate(erosion,kernel,iterations = 2)
plt.imshow(dilation)
plt.xticks([]), plt.yticks([])
plt.show()
```

```
# new image 
open = cv2.imread('/content/image.jpg',0)
plt.imshow(open)
plt.show()
```

```
opening = cv2.morphologyEx(open, cv2.MORPH_OPEN, kernel)
```

```
plt.imshow(opening)
plt.xticks([]), plt.yticks([])
plt.show()
```

```
close = cv2.imread('/content/image.jpg',0)
plt.imshow(close)
plt.show()
```

### **Closing**

```
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```
plt.imshow(closing)
plt.xticks([]), plt.yticks([])
plt.show()
```

### **Morphological Gradient**

```
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```
plt.imshow(gradient)
plt.xticks([]), plt.yticks([])
plt.show()
```

### **Top Hat**

```
ker = np.ones((7,7), np.uint8)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, ker)
```

```
plt.imshow(tophat, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

### **Black Hat**

```
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```
plt.imshow(blackhat, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```
