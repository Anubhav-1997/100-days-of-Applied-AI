```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
img = cv2.imread('/Errichto.jpg',0);
imgplot = plt.imshow(img)
plt.show()
```

```
import random
def Add_SPN(img,p):
  '''
  Objective : Add Salt and peper noise to the image
  Input : Image : img, Probability : p
  Output : image 
  '''
  output_img = np.zeros(img.shape,np.uint8)
  threhold = 1 - p
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      rand = random.random()
      if rand < p:
        output_img[i][j] = 0
      elif rand > threhold:
        output_img[i][j] = 255
      else:
        output_img[i][j] = img[i][j]
  
  return output_img


noisy_image = Add_SPN(img,0.1)
```

```
imgplot = plt.imshow(noisy_image)
plt.show()
```
